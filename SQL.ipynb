```python
#借りた日のMAX値を断酒つしてみる
sql_query = ''' SELECT customer_id,max(rental_date) as date,count(customer_id) as count FROM rental group by customer_id'''

# Pandasを利用します。
df = pd.read_sql(sql_query, con)

df.head(10)
```

```python
sql_query = ''' 
SELECT 
    A.customer_id
    ,A.email
    ,B.date
    ,B.count
    
FROM
    customer A

INNER JOIN
    (SELECT customer_id,max(rental_date) as date,count(customer_id) as count FROM rental group by customer_id) as B

ON
    A.customer_id=B.customer_id

'''

# Pandasを利用します。
df = pd.read_sql(sql_query, con)

print(len(df))
df.head(10)
```
