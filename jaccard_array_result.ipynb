```python
import pandas as pd
import numpy as np
from scipy import sparse
from scipy.spatial.distance import jaccard
```

```python
df=pd.read_csv('BOOK1.csv',encoding='shift-jis')
df
```

```python
df=df.pivot_table('CATEGOLY',index='USER',columns='PRODUCT',aggfunc='count')
df
```

```python
df=df.fillna(0)
df
```

```python
result=pd.DataFrame(index=df.index,columns=df.index)
for i in np.arange(len(df)):
    x=np.array(df.ix[i])
    u1=df.index[i]
    for j in np.arange(len(df)):
        if i==j:
            0
        else:
            u2=df.index[j]
            y=np.array(df.ix[j])
            z=jaccard(x,y)
            result.ix[u1,u2]=z
```
