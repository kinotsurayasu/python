```python
import pandas as pd
import numpy as np
from scipy import sparse
from scipy.spatial.distance import jaccard
```

```python
df=pd.read_csv('BOOK1.csv',encoding='shift-jis')
df
```

```python
df1=df.pivot_table('CATEGOLY',index='USER',columns='PRODUCT',aggfunc='count')
df1
```

```python
df1=df1.fillna(0)
df1
```

```python
result=pd.DataFrame(index=df1.index,columns=df1.index)
for i in np.arange(len(df1)):
    x=np.array(df1.ix[i])
    u1=df1.index[i]
    for j in np.arange(len(df1)):
        if i==j:
            0
        else:
            u2=df1.index[j]
            y=np.array(df1.ix[j])
            z=jaccard(x,y)
            result.ix[u1,u2]=z
```
